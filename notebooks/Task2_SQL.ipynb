```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("../data/processed/cleaned_supermarket_sales.csv")
df.head()
```

```python
conn = sqlite3.connect("../database/sales.db")
```

```python
df.to_sql("sales", conn, if_exists="replace", index=False)

print("Database Created Successfully")
```

```python
pd.read_sql("SELECT * FROM sales LIMIT 5;", conn)
```

```python
query = "SELECT * FROM sales;"
result = pd.read_sql(query, conn)
result.head()
```

```python
query = """
SELECT *
FROM sales
WHERE Gender = 'Female';
"""

result = pd.read_sql(query, conn)
result.head()
```

```python
query = """
SELECT *
FROM sales
ORDER BY Total DESC;
"""

result = pd.read_sql(query, conn)
result.head()
```

```python
query = """
SELECT *
FROM sales
ORDER BY Sales DESC;
"""

result = pd.read_sql(query, conn)
result.head()
```

```python
query = """
SELECT *
FROM sales
LIMIT 10;
"""

result = pd.read_sql(query, conn)
result
```

```python
query = """
SELECT COUNT(*) AS Total_Records
FROM sales;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT SUM(Sales) AS Total_Sales
FROM sales;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT AVG(Sales) AS Average_Sales
FROM sales;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT MIN(Sales) AS Minimum_Sales
FROM sales;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT MAX(Sales) AS Maximum_Sales
FROM sales;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT "Product line",
SUM(Sales) AS Total_Sales
FROM sales
GROUP BY "Product line";
"""

result = pd.read_sql(query, conn)
result
```

```python
query = """
SELECT "Product line",
SUM(Sales) AS Total_Sales
FROM sales
GROUP BY "Product line"
HAVING SUM(Sales) > 50000;
"""

result = pd.read_sql(query, conn)
result
```

```python
query = """
SELECT "Product line",
AVG(Sales) AS Average_Sales
FROM sales
GROUP BY "Product line";
"""

result = pd.read_sql(query, conn)
result
```

```python
query = """
SELECT City,
SUM(Sales) AS Total_Sales
FROM sales
GROUP BY City
ORDER BY Total_Sales DESC;
"""

result = pd.read_sql(query, conn)
result
```

```python
query = """
SELECT Payment,
COUNT(*) AS Total_Customers
FROM sales
GROUP BY Payment;
"""

result = pd.read_sql(query, conn)
result
```

```python
query = """
WITH AvgSales AS (
    SELECT AVG(Sales) AS avg_sales
    FROM sales
)

SELECT *
FROM sales, AvgSales
WHERE Sales > avg_sales;
"""

result = pd.read_sql(query, conn)
result.head()
```

```python
query = """
SELECT
    "Invoice ID",
    Sales,
    RANK() OVER (ORDER BY Sales DESC) AS Sales_Rank
FROM sales;
"""

result = pd.read_sql(query, conn)
result.head(10)
```

```python
query = """
CREATE VIEW HighSales AS
SELECT *
FROM sales
WHERE Sales > 500;
"""

conn.execute(query)
print("View Created Successfully!")
```

```python
query = "SELECT * FROM HighSales;"

result = pd.read_sql(query, conn)
result.head()
```

```python
query = """
SELECT City,
SUM(Sales) AS Total_Sales
FROM sales
GROUP BY City;
"""

city_sales = pd.read_sql(query, conn)

print(city_sales)
```

```python
city_sales.to_csv("../data/processed/city_sales.csv", index=False)

print("CSV Saved Successfully!")
```

```python
conn.close()

print("Database Closed Successfully!")
```
